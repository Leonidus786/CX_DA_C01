```python
# While 

## The while loop continues to execute as long as the condition is true.

count = 0


while count < 11:
    print(count)
    count = count + 1
```

```python
# Loop Control statements

# break -- it will exits the loop prematurely.

# continue -- it will skips the current iteration and continue with the next.

# pass -- it is a null operation; it does nothing.


for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

```python
for i in range(5):
    if i == 3:
        continue
    print(i)
```

```python
for i in range(5):
    if i == 3:
        break
    print(i)
```

```python
for i in range(5):
    if i == 3:
        pass
    print(i)
```

```python
for i in range(100):
    pass
```

```python
# Example = calculate the sum of first N natural numbers.
# using a while loop and for loop

## while loop 

n = 20
sum = 0
count = 1

while count <= n:
    sum = sum + count
    count += 1

print("Sum of first 20 natural numbers: ", sum)
```

```python

sum = 0

for i in range(21):
    sum = sum + i

print("Sum of first 20 natural numbers: ", sum)
```
