# Function Names
## Function names follow the same rules as variable names in Python:

- A function name must start with a letter or underscore
- A function name can only contain letters, numbers, and underscores
- Function names are case-sensitive (myFunction and myfunction are different)

## Arbitrary Arguments, \*args

- If you do not know how many arguments that will be passed into your function, add a \* (asterisk) before the parameter name in the function definition.

- This way the function will receive a tuple of arguments, and can access the items accordingly:


```python
#  creating a function with *args
def my_function(*kids):
  print(f"The youngest child is {kids[2]}." )


# calling a function with multiple variables 
my_function("Emil", "Tobias", "Linus", 'Rishi')
```

```python
## Default Parameters

def greet(name):
    print(f"Hello {name} Welcome To the paradise.")



# greet()
greet("Abhishek")
```

```python
def greet(name="Guest"):
    print(f"Hello {name} Welcome To the paradise")

# greet("Amit")
greet()
```

Arbitrary Keyword Arguments - **kwargs
If you do not know how many keyword arguments will be passed into your function, add two asterisks ** before the parameter name.

This way, the function will receive a dictionary of arguments and can access the items accordingly

```python
def my_function(**kid):
  print("His last name is " + kid["lname"])

my_function(fname = "Tobias", lname = "Refsnes")
```

```python
### Variable Length Arguments
## Positional And Keywords arguments

def print_numbers(*Amit):
    for number in Amit:
        print(number)
```

```python
print_numbers(1,2,3,4,5,"Amit",6,7,8,9,10)
```

```python
## Positional arguments
def print_numbers(*args):
    for number in args:
        print(number)
```

```python
print_numbers("Abhishek",1,2,3,4,5,6,7,8)
```

```python
### Keywords Arguments

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(name="Abhishek",age="31", city="Gurugram", country="India")
```

```python
def print_details(*args,**kwargs):
    for val in args:
        print(f" Positional arument :{val}")
    
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(1,2,3,4,"Amit", name="Amit", age="31", country="India")
```
