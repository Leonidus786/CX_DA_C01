### Functions - A block of code written to perform a specific/special task.
A function is a block of code which only runs when it is called.

### From a function's perspective:

## Parameter:- A parameter is the variable listed inside the parentheses in the function definition.

## Arguments:- An argument is the actual value that is sent to the function when it is called.

```python
a = 15
```

```python
b = 16
```

```python
c = 10
```

```python
a + b +c
```

```python
def addition(a,b,c):
    return a+b+c
```

```python
addition(10,5,10)
```

```python
addition(2,3,5)
```

```python
## Default parameter

def hello(name="Naina"):
    print(f"Hi {name}, welcome to my python course.")
```

```python
hello("Gaurav")
```

```python
hello()
```

## The phrase Keyword Arguments is often shortened to kwargs in Python documentation.

```python
def my_function(animal, name):
  print("I have a", animal)
  print("My", animal + "'s name is", name)
```

```python
my_function(animal = "dog", name = "Buddy")
```
